## Split variables into categories by creating new columns

Split into categories the following variables by creating new columns:
'Previous qualification', 'Mother's qualification', 'Father's qualification', 'Mother's occupation', 'Father's occupation'.

```python
# 'Father's occupation' column split

# Dictionary for mapping categories:
categories = {
    'Father student': [0],
    'Father managers and executives': [1],
    'Father intellectual and scientific professionals': [2, 122, 123, 125],
    'Father intermediate technicians and professionals': [3, 131, 132, 134],
    'Father administrative and support staff': [4, 141, 143, 144],
    'Father in personal service': [5, 151, 152, 153],
    'Father agricultural, forestry, and fishery workers': [6],
    'Father in industry, construction workers, and craftsmen': [7, 171, 173, 175],
    'Father machine operators and assemblers': [8],
    'Father unskilled workers': [9, 191, 192, 193, 194, 195],
    'Father in armed forces': [10],
    'Father other situations and unknown': [90, 99]
}

# new columns Initialization 
for category in categories.keys():
    df[category] = 0

# fill new columns
for category, values in categories.items():
    df.loc[df['Father\'s occupation'].isin(values), category] = 1
```

```python
# 'Mother's occupation' column split

# Dictionary for mapping categories:
categories = {
    'Mother student': [0],
    'Mother managers and executives': [1],
    'Mother intellectual and scientific professionals': [2, 122, 123, 125],
    'Mother intermediate technicians and professionals': [3, 131, 132, 134],
    'Mother administrative and support staff': [4, 141, 143, 144],
    'Mother in personal service': [5, 151, 152, 153],
    'Mother agricultural, forestry, and fishery workers': [6],
    'Mother in industry, construction workers, and craftsmen': [7, 171, 173, 175],
    'Mother machine operators and assemblers': [8],
    'Mother unskilled workers': [9, 191, 192, 193, 194],
    'Mother in armed forces': [10],
    'Mother other situations and unknown': [90, 99]
}

# new columns Initialization 
for category in categories.keys():
    df[category] = 0

# fill new columns
for category, values in categories.items():
    df.loc[df['Mother\'s occupation'].isin(values), category] = 1
```

```python
# 'Mother's qualification' column split

# Dictionary for mapping categories:
categories = {
    'Mother secondary education': [1, 9, 10, 11, 12, 14, 18, 19, 26, 27, 29, 30, 38],
    'Mother higher education': [2, 3, 4, 5, 6, 40, 41, 42, 43, 44],
    'Mother basic education': [19, 37],
    'Mother technical education': [22, 39],
    'Mother no education': [34, 35, 36],
    'Mother unknown education': [34]
}

# new columns Initialization 
for category in categories.keys():
    df[category] = 0

# fill new columns
for category, values in categories.items():
    df.loc[df['Mother\'s qualification'].isin(values), category] = 1
```

```python
# 'Father's qualification' column split

# Dictionary for mapping categories:
categories = {
    'Father secondary education': [1, 9, 10, 11, 12, 13, 14, 18, 19, 20, 25, 26, 27, 29, 30, 31, 33, 38],
    'Father higher education': [2, 3, 4, 5, 6, 40, 41, 42, 43, 44],
    'Father basic education': [19, 37],
    'Father technical education': [22, 39],
    'Father no education': [34, 35, 36],
    'Father unknown education': [34]
}

# new columns Initialization 
for category in categories.keys():
    df[category] = 0

# fill new columns
for category, values in categories.items():
    df.loc[df['Father\'s qualification'].isin(values), category] = 1
```

```python
# 'Previous qualification' column split

# Dictionary for mapping categories
categories = {
    'Previous qualification: Secondary': [1, 9, 10, 12, 14, 15],
    'Previous qualification: Higher': [2, 3, 4, 5, 6, 40, 43],
    'Previous qualification: Basic': [19, 38],
    'Previous qualification: Technical course': [39, 42]
}

# Initialize new columns to 0
for category in categories.keys():
    df[category] = 0

# Populate the new columns based on the values in the original column
for category, values in categories.items():
    df.loc[df['Previous qualification'].isin(values), category] = 1
```

```python
# Drop the original columns before split

df.drop(['Previous qualification', 'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation', 
         'Father\'s occupation'], axis=1, inplace=True)
```

```python
# Move the 'Target' column to the end of the dataframe 

target_column = df.pop('Target')
df['Target'] = target_column
```

# Correlations
<br>
We chose to display the correlation matrix after splitting the columns to obtain more precise results. This approach allows us to better understand the relationships between the variables. We will continue with data transformation and apply normalization at a later stage to avoid affecting the correlation representation.

```python
# One-hot encoding on the "Target" column for correlations visualization
df_corr = df.copy()
df_corr = pd.get_dummies(df_corr, columns = ['Target'])
```

```python
df_corr.head()
```

```python
# The blue/red color indicating if the correlation is positive or negative.

corr = df_corr.corr()
corr.style.background_gradient(cmap='RdBu_r').format("{:.2f}")
```

```python
# We can see better the whole picture using seaborn and removing the upper half of the matrix


# Calculate pairwise-correlation
matrix = df_corr.corr()

# Create a mask
mask = np.triu(np.ones_like(matrix, dtype=bool))


# Create a custom divergin palette
cmap = sns.diverging_palette(250, 15, s=75, l=40,
                            n=9, center="light", as_cmap=True)

plt.figure(figsize=(25, 16))
sns.heatmap(matrix, mask=mask, center=0, 
            square=True, cmap=cmap)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

In line with observations from the Pandas profiling report, notable correlations exist among various variables, which are logically expected. For instance, 'Daytime/evening attendance' correlates with personal situations such as 'Marital status' (married/single) and specific courses (e.g., Course_8014 and Course_9991), influencing attendance times.  <br> 

A striking example is the strong correlation (0.91) between 'Application mode 44' (Technological specialization diploma holder) and 'Previous qualification: Technical course'. This high correlation suggests that individuals with a technical background are more likely to apply for technological specializations.

### Correlations with the Target
Now focusing on the **'Target'** variable, which classifies students into three categories: dropout, enrolled, and graduate at the end of the normal course duration, key correlations of interest (â‰¥ 0.3) are observed:
<br><br>
The highest correlations with our target are found with the **Curricular variables** as follows:
<br><br>
**Curricular units 2nd sem (approved)**: 0.7 with Target_Graduate (-0.73 with Target_Dropout)<br>
This strong correlation indicates that students who pass their curricular units in the second semester are more likely to graduate, highlighting approval of these units as a robust predictor of course completion.
<br><br>
**Curricular units 2nd sem (grade)**: -0.74 with Target_Dropout<br>
The negative correlation with Target_Dropout suggests that higher grades in the second semester decrease dropout likelihood. Poor performance in this semester poses a significant dropout risk.
<br><br>
**Curricular units 1st sem (approved)**: -0.69 with Target_Dropout (0.64 with Target_Graduate)<br>
Similarly, approval of curricular units in the first semester strongly predicts graduation, underscoring early academic success as critical for course completion.
<br><br>
**Curricular units 1st sem (grade)**: -0.68 with Target_Dropout<br>
This negative correlation indicates that strong performance in the first semester reduces dropout rates, emphasizing the importance of early academic achievements.
<br><br>
**Curricular units 2nd sem (evaluations)**: -0.32 with Target_Dropout<br>
The quantity of evaluations in the second semester shows a negative correlation with dropout rates.
<br><br><br>
**Tuition fees up to date**: -0.45 with Target_Dropout<br>
Timely payment of tuition fees negatively correlates with dropout, suggesting financial stability as a key factor in course completion.
<br><br>
**Scholarship holder**: 0.4 with Target_Graduate<br>
Being a scholarship holder positively correlates with graduation, indicating that financial support enhances course completion likelihood.
<br><br>
**Application mode_39 (Over 23 years old)**: 0.35 with Target_Dropout<br>
Students over 23 years old applying under mode 39 are less likely to graduate, possibly indicating unique challenges or motivations among older students.
<br><br>
**Age at enrollment**: 0.33 with Target_Dropout<br>
Older age at enrollment correlates with higher dropout rates.
<br><br>
**Gender**: 0.3 with Target_Graduate <br>
Gender also shows a correlation with graduation, with female students less likely to graduate, suggesting gender-related disparities in academic outcomes.
<br><br>
#### Preliminary conclusions:
As expected, approved curricular units demonstrate significant correlations with dropout rates, enrollment, and course completion, particularly in the second semester. Additionally, maintaining up-to-date tuition fee payments and holding scholarships exhibit strong positive correlations with graduation. Lastly, 'Age at enrollment' and 'Gender' show notable correlations, indicating demographic factors also influence student outcomes. <br>
